```python
import cv2
from ultralytics import YOLO
from pycuber import Cube
import numpy as np
import os
import matplotlib.pyplot as plt

# Load YOLO model
model = YOLO("runs/detect/train/weights/best.pt")

def detect_rubiks_colors(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    color_masks = {}

    color_ranges = {
        "red": [([0, 120, 80], [10, 255, 255]), ([170, 120, 80], [180, 255, 255])],
        "orange": [([10, 120, 100], [22, 255, 255])],
        "yellow": [([22, 100, 100], [40, 255, 255])],
        "green": [([40, 50, 50], [80, 255, 255])],
        "blue": [([80, 50, 50], [130, 255, 255])],
        "white": [([0, 0, 180], [180, 50, 255])]
    }

    for color, ranges in color_ranges.items():
        mask = None
        for lower, upper in ranges:
            lower = np.array(lower, dtype="uint8")
            upper = np.array(upper, dtype="uint8")
            if mask is None:
                mask = cv2.inRange(hsv, lower, upper)
            else:
                mask = cv2.bitwise_or(mask, cv2.inRange(hsv, lower, upper))
        color_masks[color] = mask

    kernel = np.ones((5, 5), np.uint8)
    for color, mask in color_masks.items():
        color_masks[color] = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    total_mask = None
    for mask in color_masks.values():
        if total_mask is None:
            total_mask = mask.copy()
        else:
            total_mask = cv2.bitwise_or(total_mask, mask)

    kernel = np.ones((5, 5), np.uint8)
    total_mask = cv2.morphologyEx(total_mask, cv2.MORPH_CLOSE, kernel)

    face_colors = [["" for _ in range(3)] for _ in range(3)]
    contours_by_position = []

    for color, mask in color_masks.items():
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for contour in contours:
            area = cv2.contourArea(contour)
            if area > 100:
                x, y, w, h = cv2.boundingRect(contour)
                aspect_ratio = float(w) / h
                hull = cv2.convexHull(contour)
                hull_area = cv2.contourArea(hull)
                solidity = float(area) / hull_area if hull_area > 0 else 0
                if 0.8 <= aspect_ratio <= 1.2 and solidity > 0.8 and w > 20 and h > 20:
                    cx = x + w / 2
                    cy = y + h / 2
                    contours_by_position.append((cx, cy, color, w, h))

    contours_by_position.sort(key=lambda item: item[1])

    rows = []
    current_row = []
    last_cy = -1
    tolerance = 20

    for cx, cy, color, w, h in contours_by_position:
        if last_cy == -1 or abs(cy - last_cy) <= tolerance:
            current_row.append((cx, cy, color))
        else:
            rows.append(sorted(current_row, key=lambda item: item[0]))
            current_row = [(cx, cy, color)]
        last_cy = cy

    if current_row:
        rows.append(sorted(current_row, key=lambda item: item[0]))

    if len(rows) != 3:
        print(f"Warning: Expected 3 rows, but found {len(rows)}")
        return None, [], None

    for row_index, row in enumerate(rows):
        for col_index, (_, _, color) in enumerate(row):
            face_colors[row_index][col_index] = color

    return face_colors, contours_by_position, total_mask

def process_cube_images(folder_path):
    cube_state = {}
    face_names = ["U", "D", "F", "B", "L", "R"]

    for face_name in face_names:
        print(face_name)
        image_path = os.path.join(folder_path, f"{face_name}.jpeg")
        if not os.path.exists(image_path):
            print(f"Error: File {image_path} not found!")
            return None

        image = cv2.imread(image_path)
        if image is None:
            print(f"Error: Could not read {image_path}")
            return None

        results = model.predict(image)
        best_box = None
        highest_conf = 0.0

        for r in results:
            for box in r.boxes:
                conf = float(box.conf[0])
                if conf > highest_conf:
                    highest_conf = conf
                    best_box = box.xyxy[0]

        if best_box is not None:
            x1, y1, x2, y2 = map(int, best_box)
            margin = 0
            y1 = max(0, y1 - margin)
            y2 = min(image.shape[0], y2 + margin)
            x1 = max(0, x1 - margin)
            x2 = min(image.shape[1], x2 + margin)
            cropped = image[y1:y2, x1:x2]

            face_colors, contours_by_position, total_mask = detect_rubiks_colors(cropped)
            if face_colors is None:
                print(f"Error: Could not detect 9 colors on {face_name} face")
                return None

            fig, axs = plt.subplots(1, 4, figsize=(20, 5))  # Added one more subplot
            axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) #original full image
            axs[0].set_title("Original Image")
            axs[0].axis('off')

            axs[1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
            axs[1].set_title("Cropped Image")
            axs[1].axis('off')

            axs[2].imshow(total_mask, cmap='gray')
            axs[2].set_title("Total Mask")
            axs[2].axis('off')

            axs[3].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
            for cx, cy, color, w, h in contours_by_position:
                axs[3].add_patch(plt.Rectangle((cx - w / 2, cy - h / 2), w, h, linewidth=2, edgecolor='yellow', facecolor='none'))
                axs[3].text(cx, cy, color, fontsize=8, color='red', ha='center', va='center')
            axs[3].set_title(f'{face_name} face detection')
            axs[3].axis('off')

            plt.show()

            cube_state[face_name] = face_colors
        else:
            print(f"No Rubik's Cube detected in {face_name}.jpeg")
            return None

    return cube_state

def cube_state_to_string(cube_state):
    # Color mapping from color names to the required letters
    color_map = {"red": "R", "orange": "O", "yellow": "Y", "green": "G", "blue": "B", "white": "W"}
    
    # Define the face order: Up, Right, Front, Down, Left, Back
    face_order = ["U", "R", "F", "D", "L", "B"]
    
    # Initialize the string to store the cube state in the required format
    cube_string = ""
    
    # Iterate over the faces in the correct order
    for face_name in face_order:
        # For each face, extract the 3x3 grid and convert it to a string
        for row in range(3):
            for col in range(3):
                color = cube_state.get(face_name, [["" for _ in range(3)] for _ in range(3)])[row][col]
                
                # Map the color name to the appropriate letter
                if color in color_map:
                    cube_string += color_map[color]
                else:
                    print(f"Error: Color '{color}' on face {face_name} at position ({row}, {col}) is not recognized.")
                    return None  # Return None to indicate an error in color detection
    
    return cube_string


folder_path = r"C:\Users\lsant\Downloads\solve"  # Update with your folder path
cube_state = process_cube_images(folder_path)
if cube_state:
    cube_string = cube_state_to_string(cube_state)
    if cube_string:
        print("Cube state as string:", cube_string)
    else:
        print("Error converting cube state to string.")
else:
    print("Cube state could not be extracted.")
    
      # You can remove this print if not needed

c = cube_state_to_string(cube_state)
def replace_colors(input_string):
    replacements = {
        'W': 'U',
        'B': 'F',
        'O': 'X',
        'R': 'L',
        'Y': 'D',
        'G': 'B'
    }

    for old, new in replacements.items():
        input_string = input_string.replace(old, new)
    
    return input_string

d =  replace_colors(c)
d = d.replace('X','R')
print(d)
```
